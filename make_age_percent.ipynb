```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df=pd.read_csv("/Users/jeon-eunkyu/PycharmProjects/uul/excelFile/people/경기인구.csv",encoding="CP949") # 파일명 바꾸기
```

```python
#v1
arr = list(df.columns)

for i in range(1,len(arr)) :
    arr[i] = arr[i][11:]
    
print(arr)

df.columns = arr

df.head()
```

```python
df['sum_60'] = df.iloc[:, 9:14].sum(axis=1)
df['sum_0_20'] = df.iloc[:, 3:5].sum(axis=1)
df['sum_20_60'] = df.iloc[:, 5:9].sum(axis=1)

df['oldAgePercent'] = (df.iloc[:,-3]/df.iloc[:,2]) * 100
df['childAgePercent'] = (df.iloc[:,-3]/df.iloc[:,2]) * 100
df['adultAgePercent'] = (df.iloc[:,-3]/df.iloc[:,2]) * 100

```

```python
df.head()

```

```python
df['행정구역'] = df['행정구역'].str.replace('경기도 경기도 ', '')
df['행정구역'] = df['행정구역'].str.replace(r'\(.*\)', '',regex=True)

cbArr = []
df.head()
df.to_csv("경기인구v1.csv",index=False,encoding="CP949")

for i in range(3) :
    max_index = df['oldAgePercent'].idxmax()
    max_value = df.at[max_index, '행정구역'].rstrip()
    cbArr.append(max_value)
    df = df.drop(index=max_index)
    
    
print(cbArr)
```
